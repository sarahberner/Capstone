## Getting Hearthstone API

After getting the Hearthstone API from rapidapi, we want to combine all expansions into a single data frame. We also want to only look at COLLECTIBLE cards - there is a lot of information here regarding enchantments on minions and various in game text. We notice an important feature is that "collectible" should be true. 

```python
import pandas as pd
```

```python
import requests

url = 'https://omgvamp-hearthstone-v1.p.rapidapi.com/cards'

param = {
    "X-RapidAPI-Host": "omgvamp-hearthstone-v1.p.rapidapi.com",
    "X-RapidAPI-Key": "5fae6438d6mshaa0d7ebd4370f4fp143cc0jsnea48f0c12b12"
}

response = requests.get(url, headers = param)
response
```

```python
hearthstone_json = response.json()
```

Let's get a function together:

```python
expansions_list = ['Basic', 'Classic', 'Hall of Fame', 'Naxxramas', 
 'Goblins vs Gnomes', 'Blackrock Mountain', 'The Grand Tournament', 
 'The League of Explorers', 'Whispers of the Old Gods', 'One Night in Karazhan',
 'Mean Streets of Gadgetzan', "Journey to Un'Goro", 'Knights of the Frozen Throne', 
 'Kobolds & Catacombs', 'The Witchwood', 'The Boomsday Project', 
 "Rastakhan's Rumble", 'Rise of Shadows']
```

```python
df = pd.DataFrame()
for expansion in expansions_list:
    expansion_df = pd.DataFrame(hearthstone_json[expansion])
    df = pd.concat([df, expansion_df], ignore_index=True, sort = False)
df.head()
```

```python
df.to_csv('./hearthstone_df.csv')
```
